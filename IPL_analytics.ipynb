```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Most Wins

```python
temp=pd.DataFrame({"Winner": data['winner']})  #it creates a dataframe temp having one column Winner which contains data from the column winner of data
#the dataframe is made is of dictonary type 
```

```python
temp.head()
```

```python
countWins=temp.value_counts()
countWins #returns the teamwise how many wins a team have countwins is also a dictionary 
```

```python
Teams=[X[0]for X in countWins.keys()]


bar,ax=plt.subplots(figsize=(20,12))
ax=plt.pie(x=countWins,autopct="%1f%%",labels=Teams)
plt.title("MOST WINS IN IPL",fontsize=17)
plt.show()
```

### Most Wins in Eliminator

```python
sns.countplot(x='winner',hue='winner', data=data[data["eliminator"] == "Y"])
plt.title("Most Wins in Eliminator",fontsize=17)
plt.xticks(rotation=90)
plt.show()
```

### Toss Descion

```python


team = data['toss_winner'].unique()
decision_making_data = []

for element in team:
    team_bat = data[(data['toss_winner'] == element) & (data['toss_decision'] == 'bat')]
    team_fields = data[(data['toss_winner'] == element) & (data['toss_decision'] == 'field')]

    # Append to decision making data list
    decision_making_data.append({'Toss Winner': element, 'Decision': 'bat', 'Times': team_bat['toss_winner'].count()})
    decision_making_data.append({'Toss Winner': element, 'Decision': 'field', 'Times': team_fields['toss_winner'].count()})

# Create DataFrame from the accumulated data list
descisionMaking = pd.DataFrame(decision_making_data, columns=['Toss Winner', 'Decision', 'Times'])
```

```python
descisionMaking
```

```python
sns.catplot(x='Toss Winner',y='Times',hue='Decision',data=descisionMaking,kind='bar',height=5,aspect=2)
plt.xticks(rotation=90)
plt.title("Toss Decisions of Teams",fontsize=17)
plt.xlabel('Teams')
plt.ylabel('Descion')
plt.show()
```

### Famous Venues


```python
top_venues = data['venue'].value_counts()

# Create a DataFrame from the venue counts
venue_counts_df = pd.DataFrame({'Venue': top_venues.index, 'Counts': top_venues.values})

# Plotting using sns.barplot()
sns.barplot(x='Counts', y='Venue',hue='Venue', data=venue_counts_df)
plt.title("Famous Venues",fontsize=17)
plt.xlabel('number of matches')
plt.ylabel('Venue')
plt.show()
```
